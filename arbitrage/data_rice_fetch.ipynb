```python
## 日线
import pandas as pd
import os



# 获取所有期货合约
t = rqdatac.all_instruments(type='Future', market='cn', date=None)
future_symbols = t['underlying_symbol'].unique().tolist()

# 定义日期范围
start_date = '20170101'
end_date = '20241231'

output_file = 'D:\\FutureData\\ricequant\\1d_2017to2024_noadjust.h5'
# 使用 HDFStore 打开一个 HDF5 文件
with pd.HDFStore(output_file, mode='w') as store:
    # 遍历所有期货品种
    for symbol in future_symbols:
        # 获取主力合约数据
        df = rqdatac.futures.get_dominant_price(
            underlying_symbols=symbol,
            start_date=start_date,
            end_date=end_date,
            frequency='1d',
            fields=None,
            adjust_type='none'
        )

        # 如果数据不为空，则保存到 HDF5 文件
        if df is not None and not df.empty:
            # 将数据存储到 HDF5 文件中，使用品种代码作为键
            store.put(key=symbol, value=df, format='table')
            print(f'Saved {symbol} data to {output_file} under key "{symbol}"')
        else:
            print(f'No data found for {symbol}')

print(f'All data saved successfully to {output_file}')
```
